```python
# GK = 1  #ps^-1
# t_r = 1
# tau_0 = 10*t_r
# t_th = 0.1*t_r
# tau_thKK = 10*t_r 

def tau(l, tau_0):
    Tau = tau_0 / (np.abs(l)+1)
    return Tau

def N_K_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    N_K = GK*t_r*(1+t_r/t_th+t_r/tau(l, tau_0)) / ( (1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th) )
    return N_K

def N_Kprime_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    N_Kprime = GK*t_r*t_r/tau(l, tau_0) / ( (1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th) )
    return N_Kprime

def Nth_K_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    Nth_K = ( GK*t_r**2/(t_th*(1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th)) * (1+t_r/t_th+t_r/tau(l, tau_0)+(t_r/tau_thKK)*(t_r/tau(l, tau_0))/(1+t_r/tau_thKK)) ) / (1+t_r/tau_thKK-(t_r/tau_thKK)**2/(1+t_r/tau_thKK))
    return Nth_K

def Nth_Kprime_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    Nth_Kprime = ( GK*t_r**2/(t_th*(1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th)) * (t_r/tau(l, tau_0)+(t_r/tau_thKK)*(1+t_r/t_th+t_r/tau(l, tau_0))/(1+t_r/tau_thKK)) ) / (1+t_r/tau_thKK-(t_r/tau_thKK)**2/(1+t_r/tau_thKK))
    return Nth_Kprime
```

```python
def tau(l, tau_0):
    Tau = tau_0 / (np.abs(l)+1)
    return Tau

def onlyWP_fun(l, tau_0, t_th, t_f, t_r, tau_thKK, Gratio_term):
    onlyWP = 100*(  (1+t_r/t_f+t_r/t_th) / (1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))  )*(Gratio_term)
    return onlyWP

def WP_and_TH_fun(l, tau_0, t_th, t_f, t_r, tau_thKK, Gratio_term):
    WP_TH = 100*(  ((1+t_r/t_f)*(1+t_r/t_f+t_r/t_th+2*t_r/tau_thKK)) / ((1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_f+2*t_r/tau_thKK))  )*(Gratio_term)
    return WP_TH

def onlyTH_fun(l, tau_0, t_th, t_f, t_r, tau_thKK, Gratio_term):
    onlyTH = 100*(  ((1+t_r/t_f)*(1+t_r/t_f+t_r/t_th)) / ((1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_f+2*t_r/tau_thKK))  )*(Gratio_term)
    return onlyTH

def onlyTR_fun(l, tau_0, t_th, t_f, t_r, tau_thKK, Gratio_term, t_t, tau_TKK):
    onlyTR = 100*(  ((1+t_r/t_f)*(1+t_r/t_f+t_r/t_th+2*t_r/tau_thKK)) / ((1+2*t_t/tau_TKK)*(1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_f+2*t_r/tau_thKK))  )*(Gratio_term)
    return onlyTR

def All_fun(l, tau_0, t_th, t_f, t_r, tau_thKK, Gratio_term, t_t, tau_TKK):
    All    = 100*(  ((t_r/t_t)*(1+t_r/t_f+2*t_t/tau_TKK)*(1+t_r/t_f+t_r/t_th+2*t_r/tau_thKK)) / ((t_r/t_t+2*t_r/tau_TKK)*(1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_f+2*t_r/tau_thKK))  )*(Gratio_term)
    return All
```

```python
def tau(l, tau_0):
    Tau = tau_0 / (np.abs(l)+1)
    return Tau

def NWP_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK):
    NWP_K      = t_r*((1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*GK+(t_r/tau(l, tau_0))*GKprime) / ( (1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_f+t_r/t_th) )
    return NWP_K

def NWP_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK):
    NWP_Kprime = t_r*((1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*GKprime+(t_r/tau(l, tau_0))*GK) / ( (1+t_r/t_f+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_f+t_r/t_th) )
    return NWP_Kprime

def NTH_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK):
    NTH_K = (t_r**2/t_th) * (((1+t_r/t_f+t_r/tau_thKK)*(1+t_r/t_f+t_r/t_th+t_r/tau(l, tau_0))+(t_r/tau_thKK)*(t_r/tau(l, tau_0)))*GK+((1+t_r/t_f+t_r/tau_thKK)*(t_r/tau(l, tau_0))+(t_r/tau_thKK)*(1+t_r/t_f+t_r/t_th+t_r/tau(l, tau_0)))*GKprime)
    NTH_K = NTH_K / ((1+t_r/t_f+2*t_r/tau_thKK)*(1+t_r/t_f)*(1+t_r/t_f+t_r/t_th+t_r/tau(l, tau_0))*(1+t_r/t_f+t_r/t_th))
    return NTH_K

def NTH_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK):
    NTH_Kprime = (t_r**2/t_th) * (((1+t_r/t_f+t_r/tau_thKK)*(1+t_r/t_f+t_r/t_th+t_r/tau(l, tau_0))+(t_r/tau_thKK)*(t_r/tau(l, tau_0)))*GKprime+((1+t_r/t_f+t_r/tau_thKK)*(t_r/tau(l, tau_0))+(t_r/tau_thKK)*(1+t_r/t_f+t_r/t_th+t_r/tau(l, tau_0)))*GK)
    NTH_Kprime = NTH_Kprime / ((1+t_r/t_f+2*t_r/tau_thKK)*(1+t_r/t_f)*(1+t_r/t_f+t_r/t_th+t_r/tau(l, tau_0))*(1+t_r/t_f+t_r/t_th))
    return NTH_Kprime

def NTR_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK, t_t, tau_TKK):
    NWP_K      = NWP_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NWP_Kprime = NWP_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NTH_K      = NTH_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NTH_Kprime = NTH_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NTR_K = (t_t/t_f)/(1/t_t+2/tau_TKK) * ((1/t_t)*(NWP_K+NTH_K)+(1/tau_TKK)*(NWP_K+NTH_K+NWP_Kprime+NTH_Kprime))
    return NTR_K

def NTR_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK, t_t, tau_TKK):
    NWP_K      = NWP_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NWP_Kprime = NWP_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NTH_K      = NTH_K_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NTH_Kprime = NTH_Kprime_fun(l, tau_0, GK, GKprime, t_th, t_f, t_r, tau_thKK)
    NTR_Kprime = (t_t/t_f)/(1/t_t+2/tau_TKK) * ((1/t_t)*(NWP_Kprime+NTH_Kprime)+(1/tau_TKK)*(NWP_K+NTH_K+NWP_Kprime+NTH_Kprime))
    return NTR_Kprime
```
