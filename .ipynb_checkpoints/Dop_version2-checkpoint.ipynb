```python
def tau(l, tau_0):
    l = np.abs(l)
    Tau = tau_0 / (np.abs(l)+1)
    return Tau

def onlyX_fun(l, tau_0, t_th,t_r, tau_thKK):
    onlyX = 100*(  (1+t_r/t_th) / (1+t_r/t_th+2*t_r/tau(l, tau_0))  )
    return onlyX

def Both_fun(l, tau_0, t_th,t_r, tau_thKK):
    Both = 100*(  ((1+t_r/t_th+2*t_r/tau_thKK)) / ((1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+2*t_r/tau_thKK))  )
    return Both

def onlyth_fun(l, tau_0, t_th,t_r, tau_thKK):
    onlyth = 100*(  ((1+t_r/t_th)) / ((1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+2*t_r/tau_thKK))  )
    return onlyth
```

```python
fig, ax = plt.subplots(figsize=(9, 5), dpi=120)

tau_lKKmulti   = widgets.FloatSlider(value=10, min=0., max=10, step=1/10001 ,layout=widgets.Layout(width='75%'), description='tau_0_multiplier')
t_thmulti    = widgets.FloatSlider(value=0.1, min=0., max=10, step=1/10001    ,layout=widgets.Layout(width='75%'), description='t_th_multiplier')
tau_thKKmulti = widgets.FloatSlider(value=10,min=0., max=10, step=1/10001 ,layout=widgets.Layout(width='75%'), description='tau_thKK_multiplier')

t_r = 1
    
def need(tau_lKKmulti, t_thmulti  ,tau_thKKmulti):
    tau_0 = tau_lKKmulti*t_r
    print("tau_0 = "+str(format(tau_0,'.2f')))
    t_th = t_thmulti*t_r
    print("t_th = "+str(format(t_th,'.2f')))
    tau_thKK = tau_thKKmulti*t_r
    print("tau_thKK = "+str(format(tau_thKK,'.2f')))
    
    
    ax.cla()
    globals()['onlyX'] = []
    globals()['Both'] = []
    globals()['onlyth'] = []
    for l in range(0,7,1):
        globals()['onlyX'] += [onlyX_fun(l, tau_0, t_th,t_r, tau_thKK)]
        globals()['Both'] += [Both_fun(l, tau_0, t_th,t_r, tau_thKK)]
        globals()['onlyth'] += [onlyth_fun(l, tau_0, t_th,t_r, tau_thKK)]

    
    ax.plot( np.linspace(0, 6, 7), globals()['onlyX'], marker='v', linestyle='dashed', markersize=6, color='red' ,linewidth=1, zorder=2)
    ax.plot( np.linspace(0, 6, 7), globals()['Both'], marker='8', linestyle='dashed', markersize=6, color='magenta' ,linewidth=1, zorder=2)
    ax.plot( np.linspace(0, 6, 7), globals()['onlyNX'], marker='^', linestyle='dashed', markersize=6, color='blue' ,linewidth=1, zorder=2)
    for j, v in enumerate(globals()['onlyX']):
        ax.text(j, v-3.5, "%2.1f" %v, ha="center")
    for j, v in enumerate(globals()['Both']):
        ax.text(j, v+2.5, "%2.1f" %v, ha="center")
    for j, v in enumerate(globals()['onlyNX']):
        ax.text(j, v-5, "%2.1f" %v, ha="center")

    ax.grid()
    # ax.set_title('η ', fontsize=16)
    ax.set_ylabel('η$_\ell$(%)', fontsize=16, rotation=0)
    ax.set_yticks(np.arange(0, 101, 100/10))
    ax.set_yticks(np.arange(0, 101, 100/100), minor=True)
    ax.set_ylim(0, 100)
    ax.set_xlabel('$\ell$', fontsize=16)
    # ax.set_xticks(np.linspace(0, 6, 7), [' ', ' ', ' ', ' ', ' ', ' ', ' '])
    ax.legend(['η$_{1}$', 'η$_{2}$', 'η$_{3}$'], fontsize=16, title='cases ', loc='lower right')
    


        
# plt.tight_layout(pad=2)
# plt.subplots_adjust(left=0.1, right=0.9, 
#                     top=0.975, bottom=0.05, 
#                     wspace=0.4, hspace=0.028)
plt.show()

out = widgets.interactive_output(need, {'tau_lKKmulti':tau_lKKmulti, 't_thmulti':t_thmulti, 'tau_thKKmulti':tau_thKKmulti})
ui = widgets.VBox([widgets.Label('coefficient:'), widgets.VBox([tau_lKKmulti, t_thmulti, tau_thKKmulti]) ])
display(ui , out )
```

```python
# GK = 1  #ps^-1
# t_r = 1
# tau_0 = 10*t_r
# t_th = 0.1*t_r
# tau_thKK = 10*t_r 

def tau(l, tau_0):
    Tau = tau_0 / (np.abs(l)+1)
    return Tau

def N_K_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    N_K = GK*t_r*(1+t_r/t_th+t_r/tau(l, tau_0)) / ( (1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th) )
    return N_K

def N_Kprime_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    N_Kprime = GK*t_r*t_r/tau(l, tau_0) / ( (1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th) )
    return N_Kprime

def Nth_K_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    Nth_K = ( GK*t_r**2/(t_th*(1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th)) * (1+t_r/t_th+t_r/tau(l, tau_0)+(t_r/tau_thKK)*(t_r/tau(l, tau_0))/(1+t_r/tau_thKK)) ) / (1+t_r/tau_thKK-(t_r/tau_thKK)**2/(1+t_r/tau_thKK))
    return Nth_K

def Nth_Kprime_fun(l, tau_0, GK, t_th,t_r, tau_thKK):
    Nth_Kprime = ( GK*t_r**2/(t_th*(1+t_r/t_th+2*t_r/tau(l, tau_0))*(1+t_r/t_th)) * (t_r/tau(l, tau_0)+(t_r/tau_thKK)*(1+t_r/t_th+t_r/tau(l, tau_0))/(1+t_r/tau_thKK)) ) / (1+t_r/tau_thKK-(t_r/tau_thKK)**2/(1+t_r/tau_thKK))
    return Nth_Kprime
```

```python
fig, ax = plt.subplots(figsize=(9, 5), dpi=120)
tau_lKKmulti   = widgets.FloatSlider(value=10, min=0., max=10, step=1/10001 ,layout=widgets.Layout(width='75%'), description='tau_0_multiplier')
t_thmulti    = widgets.FloatSlider(value=0.1, min=0., max=10, step=1/10001    ,layout=widgets.Layout(width='75%'), description='t_th_multiplier')
tau_thKKmulti = widgets.FloatSlider(value=10,min=0., max=10, step=1/10001 ,layout=widgets.Layout(width='75%'), description='tau_thKK_multiplier')
GK=100
t_r = 1
    
def need(tau_lKKmulti, t_thmulti, tau_thKKmulti):
    
    tau_0 = tau_lKKmulti*t_r
    print("tau_0 = "+str(format(tau_0,'.2f')))
    t_th = t_thmulti*t_r
    print("t_th = "+str(format(t_th,'.2f')))
    tau_thKK = tau_thKKmulti*t_r
    print("tau_thKK = "+str(format(tau_thKK,'.2f')))
    
    ax.cla()
    globals()['N_K']  = []
    globals()['N_Kprime'] = []
    globals()['Nth_K'] = []
    globals()['Nth_Kprime'] = []
    for l in range(0,7,1):
        add_N_K  =  N_K_fun(l, tau_0, GK, t_th,t_r, tau_thKK)
        add_N_Kprime  =  N_Kprime_fun(l, tau_0, GK, t_th,t_r, tau_thKK)
        add_Nth_K  =  Nth_K_fun(l, tau_0, GK, t_th,t_r, tau_thKK)
        add_Nth_Kprime  =  Nth_Kprime_fun(l, tau_0, GK, t_th,t_r, tau_thKK)
        globals()['N_K'] += [add_N_K]
        globals()['N_Kprime'] += [add_N_Kprime]
        globals()['Nth_K'] += [add_Nth_K]
        globals()['Nth_Kprime'] += [add_Nth_Kprime]


    ax.plot( np.linspace(0, 6, 7), globals()['N_K']     , marker='v', linestyle='solid' , markersize=8, color='red' ,linewidth=1, zorder=1)
    ax.plot( np.linspace(0, 6, 7), globals()['N_Kprime'], marker='v', linestyle='solid' , markersize=8 , color='darkred' ,linewidth=1, zorder=2)
    ax.plot( np.linspace(0, 6, 7), globals()['Nth_K']     , marker='o', linestyle='dashed', markersize=6.5, color='blue' ,linewidth=1, zorder=1)
    ax.plot( np.linspace(0, 6, 7), globals()['Nth_Kprime'], marker='o', linestyle='dashed', markersize=6.5 , color='blueviolet' ,linewidth=1, zorder=2)


    for j, v in enumerate(globals()['N_K']):
        ax.text(j+0.1, v-3.75, "%2.2f" %v, ha="center")
    for j, v in enumerate(globals()['N_Kprime']):
        ax.text(j+0.1, v+1.4, "%2.2f" %v, ha="center")  
    for j, v in enumerate(globals()['Nth_K']):
        ax.text(j+0.08, v+2, "%2.1f" %v, ha="center") 
    for j, v in enumerate(globals()['Nth_Kprime']):
        ax.text(j+0.08, v+3, "%2.1f" %v, ha="center") 

    ax.grid()
    ax.legend(["$N^{X \ell}_{K}$", "$N^{X \ell}_{K'}$", "$N^{X \ell}_{K, th}$", "$N^{X \ell}_{K', th}$"], fontsize=15, loc='right')
    ax.set_ylabel('population', fontsize=16)
    ax.set_ylim(-0.1, 90)
    ax.set_yticks(np.arange(0, 10, 2), minor=True)
    ax.set_xlabel('$\ell$', fontsize=16)
    ax.set_xlim(-0.1, 6.6)
    # ax.set_xticks(np.linspace(0, 6, 7), [' ', ' ', ' ', ' ', ' ', ' ', ' '])
    ax.set_xticks(np.linspace(0, 6, 7), ['0', '1', '2', '3', '4', '5', '6'])

        
# plt.tight_layout(pad=2)
plt.subplots_adjust(left=0.1, right=0.9, 
                    top=0.975, bottom=0.05, 
                    wspace=0.4, hspace=0.05)
plt.show()

out = widgets.interactive_output(need, {'tau_lKKmulti':tau_lKKmulti, 't_thmulti':t_thmulti, 'tau_thKKmulti':tau_thKKmulti})
ui = widgets.VBox([widgets.Label('coefficient:'), widgets.VBox([tau_lKKmulti, t_thmulti, tau_thKKmulti]) ])
display(ui , out )
```
